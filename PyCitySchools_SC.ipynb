```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
#school_data_complete.dtypes
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Distric Summary data
total_students = len(school_data_complete)
total_schools = len(school_data_complete["school_name"].unique())
math_pass_count = len(school_data_complete.loc[school_data_complete["math_score"]>=70])
read_pass_count = len(school_data_complete.loc[school_data_complete["reading_score"]>=70])
total_budget = school_data_complete["budget"].sum()
avg_math_score = school_data_complete["math_score"].mean()
avg_read_score = school_data_complete["reading_score"].mean()
overall_passing_rate = ((avg_math_score +avg_reading_score)/2)
percent_pass_math = (math_pass_count/total_students)*100
percent_pass_read = (read_pass_count/total_students)*100
```

```python
#Distric Summary DataFrame
district_summary = [{"Students Passing Reading": percent_pass_read,
                     "Students Passing Math": percent_pass_math,
                     "Total Students": total_students,
                     "Total Schools": total_schools,
                     "Total Budget": total_budget,
                     "Overall Passing Rate": overall_passing_rate,
                     "Average Math Score": avg_math_score,
                     "Average Reading Score": avg_read_score}]
district_summary_df = pd.DataFrame(district_summary)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("{:,}".format)
district_summary_df["Students Passing Reading"] = district_summary_df["Students Passing Reading"].map("{:.2f}%".format)
district_summary_df["Students Passing Math"] = district_summary_df["Students Passing Math"].map("{:.2f}%".format)
district_summary_df["Overall Passing Rate"] = district_summary_df["Overall Passing Rate"].map("{:.2f}%".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.2f}%".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.2f}%".format)
district_summary_df = district_summary_df[["Total Schools", "Total Students", "Total Budget", "Students Passing Math", "Students Passing Reading", "Average Math Score", "Average Reading Score", "Overall Passing Rate"]]
district_summary_df
```

```python
#get the average math scores by school and include type and budege for each school
avg_math_scores = school_data_complete.groupby(["school_name"])["math_score"].mean()        
#find the avg reading score by each school
avg_read_scores = school_data_complete.groupby(["school_name"])["reading_score"].mean()
#get type and budget
school_type_budget_stuCount = school_data_complete.groupby(["school_name","type","budget"])["Student ID"].count()
#find number of students
#stu_count = school_data_complete["school_name"].value_counts()
#find the number of students passing math and number passing reading and group by school
stu_count_pass_math = school_data_complete.loc[school_data_complete["math_score"]>=70]
stu_count_pass_read = school_data_complete.loc[school_data_complete["reading_score"]>=70]
stu_pass_read_perSchool = stu_count_pass_read.groupby("school_name")["reading_score"].count()
stu_pass_math_perSchool = stu_count_pass_math.groupby("school_name")["math_score"].count()
avg_math_scores
```

```python
#convert to dataframes
avg_math_scores_df = pd.DataFrame(avg_math_scores)
avg_math_scores_df = avg_math_scores.reset_index(drop=False)
avg_read_scores_df = pd.DataFrame(avg_read_scores)
avg_read_scores_df = avg_read_scores.reset_index(drop=False)
school_type_budget_stuCount_df = pd.DataFrame(school_type_budget_stuCount)
school_type_budget_stuCount_df = school_type_budget_stuCount_df.reset_index(drop=False)
stu_pass_read_perSchool_df = pd.DataFrame(stu_pass_read_perSchool)
stu_pass_read_perSchool_df = stu_pass_read_perSchool_df.reset_index(drop=False) 
stu_pass_math_perSchool_df = pd.DataFrame(stu_pass_math_perSchool)
stu_pass_math_perSchool_df = stu_pass_math_perSchool_df.reset_index(drop=False)
```

```python
#merge Data Frames
summary1 = pd.merge(avg_math_scores_df, avg_read_scores_df, on="school_name")
summary2 = pd.merge(summary1, school_type_budget_stuCount_df, on="school_name")
summary3 = pd.merge(summary2, stu_pass_read_perSchool_df, on="school_name")
school_summary = pd.merge(summary3, stu_pass_math_perSchool_df, on="school_name")
#Set index to school name and rename columns
school_summary = school_summary.set_index("school_name")
school_summary = school_summary.rename(columns={"math_score_x": "Average Math Score", 
                                                "reading_score_x": "Average Reading Score", 
                                                "type": "School Type", "budget": "Total Budget", 
                                                "Student ID": "Total Students",
                                                "reading_score_y": "Total Students Passed Reading", 
                                                "math_score_y": "Total Students Passed Math"})
school_summary.head()
```

```python
#Add columns for Per student budget, percent passings, and overall
school_summary["Per Student Budget"] = school_summary["Total Budget"]/school_summary["Total Students"]
school_summary["% Passing Math"] = (school_summary["Total Students Passed Math"]/school_summary["Total Students"])*100
school_summary["% Passing Reading"] = (school_summary["Total Students Passed Reading"]/school_summary["Total Students"])*100
school_summary["% Overall Passing Rate"] = (school_summary["% Passing Math"] + school_summary["% Passing Reading"])/2
school_summary[["School Type", "Total Students", "Total Budget", "Per Student Budget", 
                "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]
```
